```python
#import
import pandas as pd
import numpy as no
import matplotlib.pyplot as plt
import seaborn as sns
import ssl
```

**0. Load seaborn dataset "tips"**

```python
import ssl 

ssl._create_default_https_context = ssl._create_unverified_context

tips = sns.load_dataset("tips")
tips.sample(5)
```

**1. What is the prefered day for eating?**

```python
sns.countplot(x = tips.day, palette = "mako")

#El sábado
```

**2. What days of the week has more men than women payers?**

```python
sns.countplot(x = tips.day, hue = tips.sex, palette =  "Set2")

#los sábados y domingos
```

**3. Do you think bill amounts change considerably depending on smoking?**

```python
sns.histplot(x = tips.tip, hue = tips.smoker, kde = True, fill  = False)


# A propinas altas (a partir de 6) prece que no existe diferencia entre fumadores y no fumadores. Sin embargo, a propinas bajas
    # la gráfica muestra que los NO fumadores dejan más propina que los fumadores. 
```

**4. Is tip correlated to total_bill? Draw a scatter plot**

```python
sns.scatterplot(x = "total_bill", y = "tip", data = tips)

#Si que esxiste una correlación, a mayor cuenta mayor es la propina que se deja 
```
